```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_ages(mu=50, sigma=13, num_samples=100, seed=42):
    np.random.seed(seed)

    sample_ages = np.random.normal(loc=mu,scale=sigma,size=num_samples)
    sample_ages = np.round(sample_ages,decimals=0)

    return sample_ages
```

```python
sample = create_ages()
```

```python
def percent_missing(df):
    percent_nan = 100 * house.isnull().sum() / len(house)
    percent_nan = percent_nan[percent_nan > 0].sort_values()

    return percent_nan
```

```python
# Dividing the basement columns into numeric and string 
bsmt_num_cols = ['Bsmt Unf SF', 'Total Bsmt SF', 'BsmtFin SF 2', 'BsmtFin SF 1', 'Bsmt Full Bath', 'Bsmt Half Bath']

house[bsmt_num_cols] = house[bsmt_num_cols].fillna(0)

bsmt_str_cols = ['Bsmt Qual', 'Bsmt Cond', 'BsmtFin Type 1', 'BsmtFin Type 2', 'Bsmt Exposure']

house[bsmt_str_cols] = house[bsmt_str_cols].fillna('None')
```

```python
percent_nan = percent_missing(house)

plt.figure(figsize=(8,4), dpi=200)
sns.barplot(x=percent_nan.index, y=percent_nan)
plt.xticks(rotation=90)
plt.ylim(0,1)
```

```python
house['Mas Vnr Type'] = house['Mas Vnr Type'].fillna("None")

house['Mas Vnr Area'] = house['Mas Vnr Area'].fillna(0)
```
